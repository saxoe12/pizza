```python
df3 = pd.read_sql_query('''SELECT *
from salaries
''', conn)
df3.head()
```

```python
employee = df3['Year'].value_counts()
employee = employee.reset_index()
employee.columns = ['Year', 'count']
fig = px.bar(employee, x='Year', y='count', title='Number of Employees per Year')
fig.show()
```

```python
df3['JobTitle'] = df3['JobTitle'].str.lower()

oldest = df3.groupby('JobTitle').apply(lambda x: x.nsmallest(1,'Year')['TotalPay'].values[0])
news = df3.groupby('JobTitle').apply(lambda x: x.nlargest(1,'Year')['TotalPay'].values[0])


salaries_progression =((news - oldest) / oldest)* 100
sort_progression = salaries_progression.sort_values(ascending=False)
sort_progression.head(20)
```

```python
fig = px.bar(sort_progression.head(20).reset_index(),
             x='JobTitle',
             y=0,
             color=0,
             title='Progression des salaires',
             labels={'JobTitle':'Poste','0':'Progression des salaires','0':'%'})

fig.update_layout(yaxis_tickformat = '%')
fig.show()
```
