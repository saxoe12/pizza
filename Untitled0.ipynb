# **visualisation**

```python
df4 = pd.read_sql_query('''SELECT year, AVG(TotalPay) as moyenne FROM Salaries
group by year
order by year desc;

''', conn)
fig = px.line(df4, x='Year', y='moyenne',title='evolution de revenue moyen par annee')
fig.show()
```

```python
df5 = pd.read_sql_query('''SELECT JobTitle, count(*) as nbjobs FROM Salaries
group by JobTitle
order by nbjobs desc
limit 10;

''', conn)
fig = px.bar(df5, x='JobTitle', y='nbjobs',title='les 10 jobs les plus demandés',labels={'JobTitle':'jobs','nbjobs':'nb jobs'})
fig.show()
```

```python
df6 = pd.read_sql_query('''SELECT JobTitle, AVG(TotalPay) as moyenne FROM Salaries
group by JobTitle
order by moyenne desc
limit 10;

''', conn)
fig = px.pie(df6, values='moyenne', names='JobTitle', title='Top 10 Jobs by Average Total Pay')
fig.show()
```
